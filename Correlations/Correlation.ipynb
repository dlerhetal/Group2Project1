```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from scipy.stats import linregress
import json
import csv
import os
pd.set_option('display.max_rows', None)
import datetime as dt
```

```python
year = ["2018","2019", "2020", "2021", "2022"]
traffic = [123473, 68851, 68347, 88549, 89506]
population = [1906794, 1933455, 1995343, 2012476, 2049076]
accidents = [34587, 34058, 21025, 18777, 17719]
```

```python
Full_Data_By_Year = pd.DataFrame({
    "Year": year,
    "Traffic Count": traffic,
    "Population": population,
    "Accidents": accidents
})
Full_Data_By_Year
```

```python
Full_Data_By_Year = Full_Data_By_Year.groupby(["Year"])
Full_Data_By_Year
```

```python
x_axis = population
y_axis = traffic
plt.plot(year, traffic, alpha=0.99, color="blue")
plt.title("Traffic By Year (2018-2022)")
plt.xlabel("Year")
plt.ylabel("Traffic Counts (in Millions)")
plt.savefig('Traffic')
plt.show()
plt.plot(year, population, alpha=0.99, color="green")
plt.title("Population By Year (2018-2022)")
plt.xlabel("Year")
plt.ylabel("Population (in Millions)")
plt.savefig('Population')
plt.show()
plt.plot(year, accidents, alpha=0.99, color="red")
plt.title("Accidents By Year (2018-2022)")
plt.xlabel("Year")
plt.ylabel("Accidents")
plt.savefig('Accidents')
plt.show()
```
