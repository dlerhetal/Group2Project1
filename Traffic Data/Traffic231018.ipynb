```python
import pandas as pd

files = ['TN121.VOL',
         'TN221.VOL',
         'TN321.VOL',
         'TN421.VOL',
         'TN521.VOL',
         'TN621.VOL',
         'TN721.VOL',
         'TN821.VOL',
         'TN921.VOL',
         'TN1021.VOL',
         'TN1121.VOL',
         'TN1221.VOL',
         'TN122.VOL',
         'TN222.VOL',
         'TN322.VOL',
         'TN422.VOL',
         'TN522.VOL',
         'TN622.VOL',
         'TN722.VOL',
         'TN822.VOL',
         'TN922.VOL',
         'TN1022.VOL',
         'TN1122.VOL',
         'TN1222.VOL']

pipe_list = []

# Loop over the files
for file in files:

    pipe_df = pd.read_csv(file, delimiter='|')

    pipe_list.append(pipe_df)
    
pipe_df = pd.concat(pipe_list)
```

```python
# Concatenate the dataframes along the rows
counts_df = pd.concat([fwf_df, pipe_df])
counts_df
```

```python
counts_df.to_csv('traffic.csv')
```
