```python
year = ["2018","2019", "2020", "2021", "2022"]
traffic = [ 113524755, 68252140, 204360572, 50401375, 63795994]
population = [1906794, 1933455, 1995343, 2012476, 2049076]
accidents = [34587, 34058, 21025, 18777, 17719]
```

```python
x_axis = population
y_axis = traffic
plt.plot(year, traffic, alpha=0.99, color="blue")
plt.title("Traffic By Year (2018-2022)")
plt.xlabel("Year")
plt.ylabel("Traffic Counts (in Millions)")
plt.savefig('Traffic_Scatter')
plt.show()
plt.plot(year, population, alpha=0.99, color="green")
plt.title("Population By Year (2018-2022)")
plt.xlabel("Year")
plt.ylabel("Population (in Millions)")
plt.savefig('Population_Scatter')
plt.show()
plt.plot(year, accidents, alpha=0.99, color="red")
plt.title("Accidents By Year (2018-2022)")
plt.xlabel("Year")
plt.ylabel("Accidents")
plt.savefig('Accidents')
plt.show()
```
